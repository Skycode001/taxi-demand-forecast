#  Прогнозирование заказов такси

Компания «Чётенькое такси» собрала исторические данные о заказах такси в аэропортах. Чтобы привлекать больше водителей в период пиковой нагрузки, нужно спрогнозировать количество заказов такси на следующий час. Постройте модель для такого предсказания.

Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

Вам нужно:

1. Загрузить данные и выполнить их ресемплирование по одному часу.
2. Проанализировать данные.
3. Обучить разные модели с различными гиперпараметрами. Сделать тестовую выборку размером 10% от исходных данных.
4. Проверить данные на тестовой выборке и сделать выводы.


Данные лежат в файле `taxi.csv`. Количество заказов находится в столбце `num_orders` (от англ. *number of orders*, «число заказов»).

### Ресемплирование и обработка выбросов

```python
# Ресемплирование с 10 минут до 1 часа (суммирование заказов)
df_hourly = df.resample('1h').sum()
print(f"После ресемплинга: {df_hourly.shape}")
print(f"Часовых интервалов: {len(df_hourly)}")
print(f"Потеряно данных: {len(df) - len(df_hourly)} строк (агрегация)")
```

```python
# Анализ распределения заказов
print("\nАнализ распределения заказов:")
print(f"Минимум: {df_hourly['num_orders'].min()}")
print(f"Максимум: {df_hourly['num_orders'].max()}")
print(f"Среднее: {df_hourly['num_orders'].mean():.1f}")
print(f"Медиана: {df_hourly['num_orders'].median():.1f}")
print(f"Стандартное отклонение: {df_hourly['num_orders'].std():.1f}")
```

```python
# Анализ выбросов
Q1 = df_hourly['num_orders'].quantile(0.25)
Q3 = df_hourly['num_orders'].quantile(0.75)
IQR = Q3 - Q1
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df_hourly[df_hourly['num_orders'] > upper_bound]
print(f"\nАнализ выбросов (метод IQR, множитель 1.5):")
print(f"Верхняя граница: {upper_bound:.1f}")
print(f"Часов-выбросов: {len(outliers_iqr)} ({len(outliers_iqr)/len(df_hourly)*100:.2f}%)")
print(f"Диапазон выбросов: [{outliers_iqr['num_orders'].min():.0f} - {outliers_iqr['num_orders'].max():.0f}]")
```

### Вывод по ресемплированию

Исходный временной ряд с шагом 10 минут (26 496 строк) агрегирован до часового интервала суммированием. Получено 4 416 часовых наблюдений. Уменьшение объёма данных связано с агрегацией, а не с удалением записей.

**Анализ распределения заказов:**
- Минимум: 0 заказов (ночные часы)
- Максимум: 462 заказа
- Среднее: 84.4 заказа
- Медиана: 78.0 заказа
- Стандартное отклонение: 45.0

Расхождение между средним и медианой указывает на правостороннюю асимметрию, что означает что есть редкие часы с экстремально высоким спросом.

**Решение по выбросам:** обработка не применялась, так как:
- Пиковые значения (в том числе аномально высокие) отражают реальный спрос
- Модели на основе деревьев устойчивы к выбросам
- Удаление или ограничение выбросов исказило бы бизнес-логику прогнозирования пиковых нагрузок

Это ночные часы и пиковые периоды (праздники, мероприятия). Модель не должна воспринимать их как шум - это реальные бизнес-пики, которые нужно предсказывать. Поэтому выбросы не удаляются и не ограничиваются: модели (CatBoost, RandomForest) устойчивы к аномалиям и способны их корректно обрабатывать.

**3. Почему линейные модели могут работать хуже**

Линейная регрессия чувствительна к асимметрии и выбросам. Метод наименьших квадратов минимизирует суммарную ошибку по всем наблюдениям. Поскольку пиковые часы редки, а основная масса значений сосредоточена ниже среднего, модель «пожертвует» точностью на пиках в пользу улучшения предсказаний на более частых, низких значениях. В результате линейная модель будет систематически занижать ночные пики и завышать утренние провалы.

Модели (RandomForest, CatBoost) не оптимизируют глобальную квадратичную ошибку таким образом - они делят пространство признаков на области и внутри каждой области дают предсказание, устойчивое к локальным аномалиям. Именно поэтому они предпочтительнее при асимметричном распределении с редкими, но значимыми пиками.

**4. Большой разброс (стандартное отклонение 45.0)**

При среднем 84.4 стандартное отклонение составляет больше половины среднего. Это говорит о высокой волатильности спроса. Модель должна улавливать не только сезонность, но и краткосрочные изменения.

**5. Нулевые значения (часы без заказов)**

Такие часы есть — преимущественно ночью (4:00–6:00). Модель не должна воспринимать их как аномалию. Это нормальное поведение ряда, связанное со спецификой времени суток. Поэтому добавлены временные признаки (hour, hour_sin, hour_cos), чтобы модель научилась различать день и ночь.

**6. Вывод для выбора модели**

Учитывая асимметрию распределения, высокую волатильность, наличие редких пиков и нулевых значений, CatBoost и RandomForest предпочтительнее линейной регрессии. Эти модели автоматически учитывают нелинейные взаимодействия (например, час * день недели) и не требуют нормализации целевой переменной.

### Создание признаков

```python
# Лаги - значения за предыдущие часы
# Зачем: заказы зависят от предыдущих часов (инерция спроса)
data = df_hourly.copy()

for lag in [1, 2, 3, 4, 5, 6, 12, 24, 48, 168]:
    data[f'lag_{lag}'] = data['num_orders'].shift(lag)
    # lag_1 = час назад, lag_24 = сутки назад, lag_168 = неделя назад

print("Добавлены лаги:", [f'lag_{l}' for l in [1,2,3,6,12,24,48,168]])
```

```python
# Проверь, сколько строк останется после создания лагов
max_lag = max([1, 2, 3, 4, 5, 6, 12, 24, 48, 72, 96, 168])
print(f"Максимальный лаг: {max_lag}")
print(f"Строк после shift: {len(data) - max_lag}")
```

```python
# Скользящая статистика - среднее и разброс за последние часы
# Зачем: сглаживают шум, показывают тренд
for window in [3, 6, 12, 24]:
    data[f'rolling_mean_{window}'] = data['num_orders'].shift(1).rolling(window).mean()
    data[f'rolling_std_{window}'] = data['num_orders'].shift(1).rolling(window).std()
    # rolling_mean_3 = среднее за последние 3 часа
    # rolling_std_6 = волатильность за последние 6 часов

print("Добавлены скользящие средние и стандартные отклонения")
```

```python
# Проверка корреляции между скользящими средними
corr_matrix = data[['rolling_mean_3', 'rolling_mean_6', 'rolling_mean_12', 'rolling_mean_24']].corr()
print(corr_matrix)
```

Несмотря на то что скользящие средние с разными окнами коррелируют,
для CatBoost это не является проблемой:
- Деревья решений нечувствительны к мультиколлинеарности
- Модель сама выберет наиболее информативное окно
- Остальные признаки просто не будут использоваться

```python
#fix

# Динамика час к часу (через лаги)
data['diff_hourly'] = data['lag_1'] - data['lag_2']

# Динамика день ко дню (через лаги)
data['diff_daily'] = data['lag_24'] - data['lag_48']

# Относительное изменение (процентное)
data['diff_hourly_pct'] = (data['lag_1'] - data['lag_2']) / (data['lag_2'] + 1)

print("Добавлены разностные признаки (без утечки целевой переменной)")
```

```python
# Экспоненциальное скользящее среднее (EWMA)

# Оставляем только один EWMA (span=6),
# так как у нас уже есть rolling_mean_6 и rolling_mean_24.
# EWMA_6 даст дополнительную информацию за счёт экспоненциальных весов.

data['ewma_6'] = data['num_orders'].ewm(span=6, adjust=False).mean().shift(1)

print("Добавлено экспоненциальное скользящее среднее (только ewma_6)")

# Проверка корреляции между EWMA и rolling_mean (только существующие столбцы)
# ewma_24 не создавали, поэтому проверяем только то, что есть
corr_check = data[['rolling_mean_6', 'ewma_6', 'rolling_mean_24']].corr()
print("\nКорреляция между скользящими средними:")
print(corr_check)
```

Несмотря на высокую корреляцию (0.96), оставляем оба признака,
так как для CatBoost это не проблема, а ewma_6 может быть полезен
как альтернативное представление тренда.
При использовании линейной модели один из них был бы удалён.

```python
# Временные признаки
# Базовые признаки (нужны для суточной и недельной сезонности)
data['hour'] = data.index.hour # 0-23 (суточный цикл)
data['dayofweek'] = data.index.dayofweek # 0-6 (недельный цикл)
data['weekend'] = (data['dayofweek'] >= 5).astype(int)  # 1 = выходной
```

**Временные признаки:**
- `hour` — для суточной сезонности
- `dayofweek` — для недельной сезонности
- `weekend` — производный от dayofweek, усиливающий эффект выходных дней

Признак `month` не добавлен, так как за 6 месяцев данных полноценную годовую сезонность выявить невозможно.

```python
# Цикличное кодирование - чтобы модель понимала, что 23ч и 0ч близки
# Зачем: обычные числа (0..23) плохи - 23 и 0 далеко, но по смыслу рядом
data['hour_sin'] = np.sin(2 * np.pi * data['hour'] / 24)
data['hour_cos'] = np.cos(2 * np.pi * data['hour'] / 24)

data['dow_sin'] = np.sin(2 * np.pi * data['dayofweek'] / 7)
data['dow_cos'] = np.cos(2 * np.pi * data['dayofweek'] / 7)

print("Добавлено циклическое кодирование времени")
```

```python
# Праздничные дни - в праздники спрос может отличаться
holidays = ['2018-01-01', '2018-01-02', '2018-01-03', '2018-01-04', '2018-01-05',
            '2018-01-07', '2018-01-08', '2018-02-23', '2018-03-08', '2018-03-09',
            '2018-05-01', '2018-05-02', '2018-05-09', '2018-06-12', '2018-11-04']
data['is_holiday'] = data.index.normalize().isin(pd.to_datetime(holidays)).astype(int)

print("Добавлен признак праздничного дня")
```

```python
# Время суток - утро/день/вечер/ночь
def get_day_part(hour):
    if 5 <= hour < 12: return 'morning'
    elif 12 <= hour < 17: return 'afternoon'
    elif 17 <= hour < 22: return 'evening'
    else: return 'night'

data['day_part'] = data['hour'].apply(get_day_part)
data = pd.get_dummies(data, columns=['day_part'], prefix='time')

print("Добавлены категории: утро, день, вечер, ночь")
```

```python
# Удаляем строки с NaN (появились из-за сдвигов)
df_features = data.dropna()

print(f"Создано признаков: {len(df_features.columns)}")
print(f"Размер данных: {df_features.shape}")
print(f"\nСписок всех признаков:\n{list(df_features.columns)}")

df_features.head()
```

## Обучение

###  Разделение данных на train/test по времени

```python
def train_test_temporal_split(data, test_size=0.1):
    # Разделение данных по времени (последние test_size% - тест)
    split_idx = int(len(data) * (1 - test_size))
    train = data.iloc[:split_idx].copy()
    test = data.iloc[split_idx:].copy()
    print(f"Train: {train.index.min()} - {train.index.max()} ({len(train)} часов)")
    print(f"Test: {test.index.min()} - {test.index.max()} ({len(test)} часов)")
    return train, test

# Определение признаков
target_col = 'num_orders'
exclude_cols = [target_col, 'hour', 'dayofweek']
feature_cols = [col for col in df_features.columns if col not in exclude_cols]

print(f"Количество признаков: {len(feature_cols)}")

# Разделение
train, test = train_test_temporal_split(df_features, test_size=0.1)

X_train = train[feature_cols]
y_train = train[target_col]
X_test = test[feature_cols]
y_test = test[target_col]

print(f"\nX_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
```

### Масштабирование признаков

```python
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Масштабирование выполнено")
print(f"Тип X_train_scaled: {type(X_train_scaled)}")
print(f"Размер: {X_train_scaled.shape}")
```

### Обучение CatBoost с оптимизацией гиперпараметров

```python
tscv = TimeSeriesSplit(n_splits=3)

# Базовая модель
catboost_base = CatBoostRegressor(random_state=42, verbose=0, thread_count=-1)

# Сетка гиперпараметров
param_grid = {
    'iterations': [100, 200, 300],
    'depth': [4, 6, 8],
    'learning_rate': [0.03, 0.05, 0.1],
    'l2_leaf_reg': [1, 3, 5]
}

# GridSearch с TimeSeriesSplit
grid_search = GridSearchCV(
    estimator=catboost_base,
    param_grid=param_grid,
    scoring='neg_root_mean_squared_error',
    cv=tscv,
    n_jobs=-1,
    verbose=1
)

grid_search.fit(X_train_scaled, y_train)

best_catboost = grid_search.best_estimator_
print(f"Лучшие параметры CatBoost: {grid_search.best_params_}")
print(f"Лучший CV RMSE: {-grid_search.best_score_:.2f}")
```

```python
# Предсказание на тесте (получаем Test RMSE)
best_catboost = grid_search.best_estimator_
pred_catboost = best_catboost.predict(X_test_scaled)
rmse_catboost = np.sqrt(mean_squared_error(y_test, pred_catboost))
print(f"Test RMSE: {rmse_catboost:.2f}")

# Сравнение
print(f"Сравнение: CV RMSE = {-grid_search.best_score_:.2f} | Test RMSE = {rmse_catboost:.2f}")
```
